```
import numpy as np
import pandas as pd
import random as rnd

import seaborn as sns
sns.set_palette('Set2')
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv('metrics.csv')
data.head()
```

```
prob = [0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9]
fig, axs = plt.subplots(len(prob), 1, figsize=(5,5*len(prob)), subplot_kw={'aspect':'equal'})

labels = ['Positions found in both vcf - TP', 'Positions found only in our vcf - FP']

i = 0 
for ax, p in zip(axs,prob):
    sizes = [data['TP1'][i], data['FP1'][i]]
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True)
    ax.set_title(p)
    i += 1
plt.show()
```

```
# Creating new dataframe out of data dataframe
lst1 = ['File Name', 'Probability', 'TP1', 'FP1', 'FN1', 'Precision1', 'Recall1', 'F1-score1']
lst2 = ['File Name', 'Probability', 'TP2', 'FP2', 'FN2', 'Precision2', 'Recall2', 'F1-score2']
lst3 = ['File Name', 'Probability', 'TP3', 'FP3', 'FN3', 'Precision3', 'Recall3', 'F1-score3']
df_1 = data.loc[:,lst1]
df_1 = df_1.rename(columns={'TP1':'TP', 'FP1':'FP', 'FN1':'FN', 'Precision1':'Precision', 'Recall1':'Recall', 'F1-score1':'F1-score'})
df_2 = data.loc[:,lst2]
df_2 = df_2.rename(columns={'TP2':'TP', 'FP2':'FP', 'FN2':'FN', 'Precision2':'Precision', 'Recall2':'Recall', 'F1-score2':'F1-score'})
df_3 = data.loc[:,lst3]
df_3 = df_3.rename(columns={'TP3':'TP', 'FP3':'FP', 'FN3':'FN', 'Precision3':'Precision', 'Recall3':'Recall', 'F1-score3':'F1-score'})

frames = [df_1,df_2,df_3]
df_new = pd.concat(frames, axis = 0, ignore_index=True)
type_one = ['First']*10
type_two = ['Second']*10
type_three = ['Third']*10
metrics_type = []
metrics_type.extend(type_one)
metrics_type.extend(type_two)
metrics_type.extend(type_three)

df_new['Metrics type'] = metrics_type
df_new
```

```
fig1, ax1 = plt.subplots(figsize=(10,5))
fig2, ax2 = plt.subplots(figsize=(10,5))
fig3, ax3 = plt.subplots(figsize=(10,5))


g1 = sns.lineplot(x="Probability", y='Precision', data=df_new.iloc[0:20,:], hue = 'Metrics type', style = 'Metrics type', markers=True, ax=ax1)
g1.set_xlabel('Probability of a success')
g1.set_title("Precision as a function of the probability of a success for two different types of calculation")
g1.figure.savefig("Precision.png")
g2 = sns.lineplot(x="Probability", y='Recall', data=df_new.iloc[0:20,:], hue = 'Metrics type', style = 'Metrics type', markers=True, ax=ax2)
g2.set_xlabel('Probability of a success')
g2.set_title("Recall as a function of the probability of a success for two different types of calculation")
g2.figure.savefig("Recall.png")
g3 = sns.lineplot(x="Probability", y='F1-score', data=df_new.iloc[0:20,:], hue = 'Metrics type', style = 'Metrics type', markers=True, ax=ax3)
g3.set_xlabel('Probability of a success')
g3.set_title("F1-score as a function of the probability of a success for two different types of calculation")
g3.figure.savefig("F1-score.png")
```

```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['0.5' , '0.55', '0.6' , '0.65', '0.7' , '0.75', '0.8' , '0.85', '0.9' , '0.95']
TP_1 = data['TP2'].values
FP_1 = data['FP2'].values
FN_1 = data['FN2'].values

x = np.arange(len(labels))  # the label locations
width = 0.20  # the width of the bars

fig, ax = plt.subplots(figsize=(20,5))
rects1 = ax.bar(x, TP_1, width, label='TP')
rects2 = ax.bar(x + width, FP_1, width, label='FP')
rects3 = ax.bar(x - width, FN_1, width, label='FN')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Probability of a success')
ax.set_ylabel('Value')
ax.set_title('Values of TP, FP and FN depending on the probability of a success')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
fig.tight_layout()
fig.savefig("II-TPFPFN-bar.png")
plt.show()
```
