```
import numpy as np
import pandas as pd
import random as rnd

import seaborn as sns
sns.set_palette('Set2')
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv('metrics.csv')
data.head()
```

```
# pie chart of workers
labels = ['Positions found in both vcf - TP', 'Positions found only in our vcf - FP']

sizes = [data['TP1'][0], data['FP1'][0]]
# print(sizes) # adds up to 1433, which is the total number of participants
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True)
ax1.axis('equal')
ax1.figure.savefig("TP1.png")
plt.show()
```

```
prob = [0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9]
fig, axs = plt.subplots(len(prob), 1, figsize=(5,5*len(prob)), subplot_kw={'aspect':'equal'})

labels = ['Positions found in both vcf - TP', 'Positions found only in our vcf - FP']

i = 0 
for ax, p in zip(axs,prob):
    sizes = [data['TP1'][i], data['FP1'][i]]
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True)
    ax.set_title(p)
    i += 1
plt.show()
```

```
# Creating new dataframe out of data dataframe
lst1 = ['File Name', 'Probability', 'TP1', 'FP1', 'FN1', 'Precision1', 'Recall1', 'F1-score1']
lst2 = ['File Name', 'Probability', 'TP2', 'FP2', 'FN2', 'Precision2', 'Recall2', 'F1-score2']
lst3 = ['File Name', 'Probability', 'TP3', 'FP3', 'FN3', 'Precision3', 'Recall3', 'F1-score3']
df_1 = data.loc[:,lst1]
df_1 = df_1.rename(columns={'TP1':'TP', 'FP1':'FP', 'FN1':'FN', 'Precision1':'Precision', 'Recall1':'Recall', 'F1-score1':'F1-score'})
df_2 = data.loc[:,lst2]
df_2 = df_2.rename(columns={'TP2':'TP', 'FP2':'FP', 'FN2':'FN', 'Precision2':'Precision', 'Recall2':'Recall', 'F1-score2':'F1-score'})
df_3 = data.loc[:,lst3]
df_3 = df_3.rename(columns={'TP3':'TP', 'FP3':'FP', 'FN3':'FN', 'Precision3':'Precision', 'Recall3':'Recall', 'F1-score3':'F1-score'})

frames = [df_1,df_2,df_3]
df_new = pd.concat(frames, axis = 0, ignore_index=True)
type_one = ['First']*10
type_two = ['Second']*10
type_three = ['Third']*10
metrics_type = []
metrics_type.extend(type_one)
metrics_type.extend(type_two)
metrics_type.extend(type_three)

df_new['Metrics type'] = metrics_type
df_new
```

```
fig, ax = plt.subplots(figsize=(10,5))

g1 = sns.lineplot(x="Probability", y="TP", data=df_new, hue = 'Metrics type', style = 'Metrics type', markers=True, ax=ax)
#g2 = sns.lineplot(x="Probability", y="FP", data=df_new, style = 'Metrics type', markers=True, ax=ax[1])
#g1 = sns.lineplot(x="Probability", y='FN', data=df_new, style = 'Metrics type', markers=True, ax=ax[2])
#g2 = sns.lineplot(x="Probability", y='Precision', data=df_new, style = 'Metrics type', markers=True, ax=ax[3])
#g1 = sns.lineplot(x="Probability", y='Recall', data=df_new, style = 'Metrics type', markers=True, ax=ax[4])
#g2 = sns.lineplot(x="Probability", y='F1-score', data=df_new, style = 'Metrics type', markers=True, ax=ax[5])
g1.figure.savefig("TP.png")
```
